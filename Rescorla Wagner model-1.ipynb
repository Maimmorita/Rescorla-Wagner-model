```python
import numpy as np
import matplotlib.pyplot as plt

w_0 = 0.0
eta = 0.5
R = [1.0] * 6 + [0.0] * 6  #US
I = [1.0] * 12  #CS
T = 12
z = 1
w = np.zeros(T+1,dtype='float')
w[0]=w_0
r = np.zeros(T,dtype='float')

for t in range (T):
 w_hat = w[t] + eta * I[t] * (R[t] - z * w[t])
 w[t+1] = w_hat
 r_hat = z * w_hat
 r[t] = r_hat

print(w)
print(r)

fig,axes = plt.subplots(2,1)

y1 = w
y2 = r

axes[0].plot(y1)
axes[0].set_ylabel('w')
axes[0].grid(True)

x2 = [I for I in range (len(r))]
axes[1].bar(x2, y2)
axes[1].set_xlabel('trial')
axes[1].set_ylabel('r')
```
